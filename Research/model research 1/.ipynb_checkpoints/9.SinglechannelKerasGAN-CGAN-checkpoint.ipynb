```python
import os
import shutil
from pathlib import Path
from PIL import Image
import logging

class CopyAndAugmentBalancedDataset:
    def __init__(self, source_directory, copy_directory, target_size=(256, 256)):
        self.source_directory = Path(source_directory)
        self.copy_directory = Path(copy_directory)
        self.target_size = target_size
        self.class_indices = self._get_class_indices()
        # Calculate min_count after obtaining all class indices to ensure it is available when needed
        self.min_count = min(len(files) for files in self.class_indices.values())
        self._log_class_info()

    def _get_class_indices(self):
        """Scan the source directory and map each class to its images."""
        class_indices = {}
        for class_name in os.listdir(self.source_directory):
            class_path = self.source_directory / class_name
            if class_path.is_dir():
                class_indices[class_name] = [str(class_path / fname) for fname in os.listdir(class_path)]
        return class_indices

    def _log_class_info(self):
        """Log information about the classes and their image counts."""
        logging.info(f"Number of unique classes: {len(self.class_indices)}")
        for class_name, files in self.class_indices.items():
            logging.info(f"Count of images in {class_name}: {len(files)}")
        logging.info(f"Minimum frequency among classes: {self.min_count}")

    def _augment_and_copy_image(self, source_path, target_path):
        """Resize image and save to target path."""
        try:
            with Image.open(source_path) as img:
                img_resized = img.resize(self.target_size, Image.Resampling.LANCZOS)
                img_resized.save(target_path)
            logging.debug(f"Image {source_path} resized and copied to {target_path}")
        except Exception as e:
            logging.error(f"Error processing image {source_path}: {e}")

    def copy_and_augment_balanced_dataset(self):
        """Copy and augment a balanced dataset to the copy directory."""
        logging.info("Starting to copy and augment the dataset...")
        for class_name, files in self.class_indices.items():
            copy_path = self.copy_directory / class_name
            copy_path.mkdir(parents=True, exist_ok=True)
            # Ensure we only process up to min_count images per class
            selected_files = files[:self.min_count]
            for file_path in selected_files:
                target_file_path = copy_path / Path(file_path).name
                self._augment_and_copy_image(file_path, target_file_path)
        logging.info(f"Augmented balanced dataset copied to {self.copy_directory}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model

# New image dimension
image_size = image_size

# Create the discriminator.
# Discriminator with deeper introduction of class labels
def create_discriminator(num_classes, ndf=64, nc=3, image_size=64):
    image_inputs = layers.Input(shape=(image_size, image_size, nc))
    labels = layers.Input(shape=(), dtype='int32')

    # Initial convolution layer reduces size by half
    x = layers.Conv2D(ndf, (4, 4), strides=(2, 2), padding='same')(image_inputs)
    x = layers.LeakyReLU(0.2)(x)

    # Second convolution layer reduces size by half again
    x = layers.Conv2D(ndf * 2, (4, 4), strides=(2, 2), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)

    # Third convolution layer reduces size by half once more
    x = layers.Conv2D(ndf * 4, (4, 4), strides=(2, 2), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)

    # Calculate the size for embedding to match the current feature map size
    current_size = image_size // 8  # Adjust based on number of convolutions above
    class_embedding = layers.Embedding(num_classes, ndf * 8 * current_size * current_size)(labels)
    class_embedding = layers.Reshape((current_size, current_size, ndf * 8))(class_embedding)
    
    # Concatenating class embeddings with feature maps
    x = layers.Concatenate()([x, class_embedding])

    # Additional convolution layer if needed or directly to final layers
    x = layers.Conv2D(ndf * 8, (4, 4), strides=(2, 2), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)

    x = layers.Flatten()(x)
    final_output = layers.Dense(1, activation='sigmoid')(x)

    return Model(inputs=[image_inputs, labels], outputs=final_output)


# Example usage for RGB and Grayscale
num_classes = num_classes
ndf = ndf
nc = num_channels  # Number of output channels; set to 1 for grayscale or 3 for RGB
image_size = image_size  # Assuming a larger image size for example
# Create an instance of the Discriminator
discriminator = create_discriminator(num_classes=num_classes, ndf=ndf, nc=nc, image_size=image_size)


generator_image_size = int(image_size/4)


# Create the generator.
import tensorflow as tf
from tensorflow.keras import layers, Model, metrics, optimizers, losses

# class ConditionalBatchNorm(layers.Layer):
#     def __init__(self, num_features, num_classes):
#         super().__init__()
#         self.num_features = num_features
#         self.bn = layers.BatchNormalization()
#         self.embed = layers.Embedding(num_classes, 2 * num_features)

#     def call(self, x, labels):
#         # Embed labels and reshape for broadcasting
#         gamma_beta = self.embed(labels)
#         gamma, beta = tf.split(gamma_beta, 2, axis=-1)
#         gamma = tf.reshape(gamma, (-1, 1, 1, self.num_features))
#         beta = tf.reshape(beta, (-1, 1, 1, self.num_features))
        
#         # Apply Conditional Batch Normalization
#         return self.bn(x) * (1 + gamma) + beta



# def create_generator(num_classes, latent_dim, ngf, num_channels, image_size):
#     # Latent inputs and class labels
#     latent_inputs = layers.Input(shape=(latent_dim,))
#     labels = layers.Input(shape=(1,), dtype='int32')

#     # Initial Dense layer
#     initial_size = image_size // 16
#     x = layers.Dense(initial_size * initial_size * ngf * 8, use_bias=False)(latent_inputs)
#     x = layers.Reshape((initial_size, initial_size, ngf * 8))(x)

#     # Convolutional layers for upscaling
#     x = layers.Conv2DTranspose(ngf * 4, (4, 4), strides=(2, 2), padding="same", use_bias=False)(x)
#     x = layers.BatchNormalization()(x)
#     x = layers.ReLU()(x)

#     x = layers.Conv2DTranspose(ngf * 2, (4, 4), strides=(2, 2), padding="same", use_bias=False)(x)
#     x = layers.BatchNormalization()(x)
#     x = layers.ReLU()(x)

#     x = layers.Conv2DTranspose(ngf, (4, 4), strides=(2, 2), padding="same", use_bias=False)(x)
#     x = ConditionalBatchNorm(ngf, num_classes)(x, labels)  # Correctly pass x and labels as separate arguments
#     x = layers.ReLU()(x)

#     # Output layer
#     outputs = layers.Conv2DTranspose(num_channels, (4, 4), strides=(2, 2), padding="same", activation='tanh')(x)

#     return Model(inputs=[latent_inputs, labels], outputs=outputs)


import tensorflow as tf
from tensorflow.keras import layers, Model



# def create_generator(num_classes, latent_dim, ngf, num_channels, image_size):
#     # Latent inputs and class labels
#     latent_inputs = layers.Input(shape=(latent_dim,))
#     labels = layers.Input(shape=(1,), dtype='int32')

#     # Initial Dense layer
#     initial_size = image_size // 16
#     x = layers.Dense(initial_size * initial_size * ngf * 8, use_bias=False)(latent_inputs)
#     x = layers.Reshape((initial_size, initial_size, ngf * 8))(x)

#     # Convolutional layers for upscaling
#     x = layers.Conv2DTranspose(ngf * 4, (4, 4), strides=(2, 2), padding="same", use_bias=False)(x)
#     x = layers.BatchNormalization()(x)
#     x = layers.ReLU()(x)

#     x = layers.Conv2DTranspose(ngf * 2, (4, 4), strides=(2, 2), padding="same", use_bias=False)(x)
#     x = layers.BatchNormalization()(x)
#     x = layers.ReLU()(x)

#     x = layers.Conv2DTranspose(ngf, (4, 4), strides=(2, 2), padding="same", use_bias=False)(x)
#     x = ConditionalBatchNorm(ngf, num_classes)(x, labels)  # Correctly pass x and labels as separate arguments
#     x = layers.ReLU()(x)

#     # Output layer
#     outputs = layers.Conv2DTranspose(num_channels, (4, 4), strides=(2, 2), padding="same", activation='tanh')(x)

#     return Model(inputs=[latent_inputs, labels], outputs=outputs)


def create_generator(num_classes, latent_dim, ngf, num_channels, image_size):
    # Inputs for latent vector and labels
    latent_inputs = layers.Input(shape=(latent_dim,))
    labels = layers.Input(shape=(1,), dtype='int32')

    # Label embedding and concatenation
    label_embedding = layers.Embedding(num_classes, 10)(labels)
    label_embedding = layers.Flatten()(label_embedding)
    merged_input = layers.Concatenate()([latent_inputs, label_embedding])

    # First dense layer
    x = layers.Dense(256, activation='relu')(merged_input)
    x = layers.LeakyReLU(0.2)(x)

    # Second dense layer
    x = layers.Dense(512, activation='relu')(x)
    x = layers.LeakyReLU(0.2)(x)

    # Third dense layer
    x = layers.Dense(1024, activation='relu')(x)
    x = layers.LeakyReLU(0.2)(x)

    # Output dense layer
    x = layers.Dense(image_size * image_size * num_channels, activation='tanh')(x)
    output = layers.Reshape((image_size, image_size, num_channels))(x)

    return Model(inputs=[latent_inputs, labels], outputs=output)


# Example usage for RGB and Grayscale
num_classes = num_classes
latent_dim = latent_dim
ngf = ngf
nc = num_channels  # Number of output channels; set to 1 for grayscale or 3 for RGB
image_size = image_size  # Assuming a larger image size for example
# generator_rgb = Generator(num_classes, latent_dim, ngf, nc, image_size)
generator_grayscale = create_generator(num_classes, latent_dim, ngf, nc, image_size)  # Grayscale example
```

```python
class ConditionalGAN(keras.Model):
    def __init__(self, discriminator, generator, latent_dim):
        super().__init__()
        self.discriminator = discriminator
        self.generator = generator
        self.latent_dim = latent_dim
        self.gen_loss_tracker = keras.metrics.Mean(name="generator_loss")
        self.disc_loss_tracker = keras.metrics.Mean(name="discriminator_loss")

    @property
    def metrics(self):
        return [self.gen_loss_tracker, self.disc_loss_tracker]

    def compile(self, d_optimizer, g_optimizer, loss_fn):
        super().compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer
        self.loss_fn = loss_fn

    def train_step(self, data):
        # Unpack data, which includes real images and their corresponding labels
        real_images, labels = data
    
        # Convert labels if necessary (e.g., if they are one-hot encoded)
        if len(labels.shape) > 1:
            labels = tf.argmax(labels, axis=-1)
    
        # Sample random points in the latent space
        batch_size = tf.shape(real_images)[0]
        random_latent_vectors = tf.random.normal(shape=(batch_size, self.latent_dim))
    
        # Training the discriminator and generator
        with tf.GradientTape() as disc_tape, tf.GradientTape() as gen_tape:
            # Generate fake images
            generated_images = self.generator([random_latent_vectors, labels], training=True)
    
            # Discriminator decisions for both real and fake images
            real_decision = self.discriminator([real_images, labels], training=True)
            fake_decision = self.discriminator([generated_images, labels], training=True)
    
            # Discriminator loss
            d_loss_real = self.loss_fn(tf.ones_like(real_decision), real_decision)
            d_loss_fake = self.loss_fn(tf.zeros_like(fake_decision), fake_decision)
            d_loss = (d_loss_real + d_loss_fake) / 2
    
            # Generator loss
            g_loss = self.loss_fn(tf.ones_like(fake_decision), fake_decision)
    
        # Calculate and apply gradients
        d_gradients = disc_tape.gradient(d_loss, self.discriminator.trainable_weights)
        g_gradients = gen_tape.gradient(g_loss, self.generator.trainable_weights)
        self.d_optimizer.apply_gradients(zip(d_gradients, self.discriminator.trainable_weights))
        self.g_optimizer.apply_gradients(zip(g_gradients, self.generator.trainable_weights))
    
        # Update and log the losses
        self.gen_loss_tracker.update_state(g_loss)
        self.disc_loss_tracker.update_state(d_loss)

        logging.info(f"Real Images shape: {real_images.shape}")
        logging.info(f"One-hot Labels shape: {labels.shape}")
        logging.info(f"Generated Images shape: {generated_images.shape}")
        logging.info(f"Generator Loss: {self.gen_loss_tracker.result()}")
        logging.info(f"Discriminator Loss: {self.disc_loss_tracker.result()}")


        return {
            "g_loss": self.gen_loss_tracker.result(),
            "d_loss": self.disc_loss_tracker.result(),
        }

# Remember to re-instantiate and recompile your model after making changes to the class definition.
```

# Interpolating between classes with the trained generator

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model

def interpolate_class(generator, first_number, second_number, num_interpolation=10, latent_dim=128, num_classes=10):
    # Create a latent space vector (noise)
    interpolation_noise = tf.random.normal(shape=(num_interpolation, latent_dim))

    # Generate interpolation vectors for labels
    linspace = np.linspace(0, 1, num_interpolation)
    # Directly determine the class index based on the interpolation threshold
    labels_interpolated = np.where(linspace < 0.5, first_number, second_number)

    # Convert labels to a TensorFlow tensor
    labels_interpolated = tf.convert_to_tensor(labels_interpolated, dtype=tf.int32)

    # Run inference with the generator
    fake_images = generator.predict([interpolation_noise, labels_interpolated])
    return fake_images

# Example usage
start_class = 0  # Starting class index
end_class = 0    # Ending class index
num_classes = 7  # Total number of classes
latent_dim = 128  # This is the correct latent dimension as expected by your generator

 # Define the directory where the models are saved
save_dir = Path(os.getcwd()) / "Model" / "GAN_CGAN_Merge"

# Load the generator model
generator_model_path = save_dir / "generator_model.keras"
generator = load_model(generator_model_path, compile=False)

# Compile the generator
generator.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0003),
    loss=keras.losses.BinaryCrossentropy(from_logits=True)
)

# Assuming `trained_gen` is your trained generator model
fake_images = interpolate_class(generator, start_class, end_class, num_interpolation=100, latent_dim=latent_dim, num_classes=num_classes)
```
